## Python Demo

```python
import requests
res = requests.get('https://www.coingecko.com/price_charts/1/usd/max.json?locale=zh-tw')
```

```python
import pandas
btc  = pandas.DataFrame(res.json()['stats'])
```

```python
btc.columns = ['tm', 'price']
```

```python
%pylab inline
btc.set_index('tm').plot(kind = 'line')
```

```python
btc = btc.set_index('tm')
```

```python
btc['sm30'] = btc['price'].rolling(window = 30).mean()
```

```python
%pylab inline
btc.plot(kind = 'line')
```

```python
btc.to_excel('btc.xlsx')
```
